```python
endpoint = 'https://api.worldbank.org/v2/country/us;can/indicator/NY.GDP.PCAP.PP.KD'

params = {
    'format': 'json', 
    'page': 1, 
    'date':'2000:2021'
}

response = requests.get(endpoint, params = params)
res = response.json()
df = pd.json_normalize(res[1])
df
```

```python
def get_all_records_json(endpoint, params, normalize = False):
    
    """
    This function takes in an endpoint to data formatted as string, and params as a dictionary (no per_pages), and 
    an option to normalize the data and will gather all of the data from that endpoint with params and turn it
    into a dataframe.
    
    
    """

    response = requests.get(endpoint, params = params)
    res = response.json()
    total_pages = res[0]['total']
    
    new_params = params
    new_params['per_page'] = total_pages
    
    response = requests.get(endpoint, params=params)
    res = response.json()
    
    if normalize:
        df = pd.json_normalize(res[1])
    else:
        df = pd.DataFrame(res[1])
    
    return df


endpoint5 = 'https://api.worldbank.org/v2/country/us;can/indicator/NY.GDP.PCAP.PP.KD'
params5 =  {
    'format': 'json',
    'date':'2000:2021'
}

get_all_records_json(endpoint5,params5, normalize = True)
    
```
